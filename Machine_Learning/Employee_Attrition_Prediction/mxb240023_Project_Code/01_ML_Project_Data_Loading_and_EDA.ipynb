```python
#import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8')   # Fix here
```

```python
#load data
df = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')
df.head()
```

```python
#explore basics
df.shape
df.info()
df.describe()
```

```python
#check Attrition distribution
df['Attrition'].value_counts()
```

```python
#EDA (Exploratory Data Analysis)
#This is where we explore patterns between features and attrition.

#Basic Visual EDA
#Plot Attrition Count (Bar Plot)
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Attrition', palette='Set2')
plt.title("Attrition Count")
plt.show()
```

```python
#Attrition Percentages
df['Attrition'].value_counts(normalize=True) * 100
```

```python
#Pie Chart (optional but looks good in your PPT)
plt.figure(figsize=(6,6))
df['Attrition'].value_counts().plot.pie(
    autopct='%1.1f%%',
    labels=['No', 'Yes'],
    colors=['skyblue', 'salmon']
)
plt.title("Attrition Distribution")
plt.ylabel("")
plt.show()
```

```python
#Numerical Feature Analysis
#Boxplot: Age vs Attrition
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x='Attrition', y='Age', palette='Set3')
plt.title("Age vs Attrition")
plt.show()
```

```python
#Boxplot: Monthly Income vs Attrition
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x='Attrition', y='MonthlyIncome', palette='Set1')
plt.title("Monthly Income vs Attrition")
plt.show()
```

```python
#Histograms for numeric features (optional)
df[['Age', 'MonthlyIncome', 'DistanceFromHome']].hist(figsize=(12,6), bins=20)
plt.show()
```

```python
#Categorical Feature Analysis
'''
We check factors like:

Job Role

Department

Business Travel

Overtime

Marital Status
'''
#Attrition by Overtime
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='OverTime', hue='Attrition', palette='coolwarm')
plt.title("Attrition vs OverTime")
plt.show()

#Attrition by Job Role
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='JobRole', hue='Attrition', palette='viridis')
plt.title("Attrition by Job Role")
plt.xticks(rotation=45)
plt.show()

#Attrition by Marital Status
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='MaritalStatus', hue='Attrition', palette='Set2')
plt.title("Attrition by Marital Status")
plt.show()
```

```python
#Correlation Heatmap (numerical only)
# Select only numeric columns
numeric_df = df.select_dtypes(include=['int64', 'float64'])

plt.figure(figsize=(14,10))
sns.heatmap(numeric_df.corr(), annot=False, cmap='coolwarm')
plt.title("Correlation Heatmap (Numerical Features Only)")
plt.show()
```
